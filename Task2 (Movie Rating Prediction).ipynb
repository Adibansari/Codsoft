# Joining Datasets

```python
merge_data = pd.merge(users_data,ratings_data,left_on="UserID",right_on="ID")
merge_data = pd.merge(merge_data,movies_data,on="MovieID")
merge_data.head(20)
```

```python
merge_data.shape
```

```python
merge_data.tail(20)
```

```python
merge_data.info()
```

```python
merge_data.isnull().sum()
```

```python
# Group the data by MovieID and Ratings, and calculate the count of users for each combination
rating_counts = merge_data.groupby(['MovieID', 'Ratings']).size().reset_index(name='UserCount')
```

```python
# Filter for movies with a certain rating count threshold (e.g., movies with at least 100 users who rated it)
threshold = 100
filtered_data = rating_counts[rating_counts['UserCount'] >= threshold]
```

```python
# Merge with the original dataset to include gender, age, occupation, and genre information
filtered_data = pd.merge(filtered_data, merge_data[['MovieID', 'Ratings', 'Gender', 'Age', 'Occupation', 'Genre']], on=['MovieID', 'Ratings'], how='inner')
```

```python
filtered_data.head(20)
```

```python
filtered_data.shape
```

# Building Model

```python
X = filtered_data[['Gender', 'Age', 'Occupation', 'Genre']]
y = filtered_data['Ratings']
```

```python
# Convert categorical variables to numerical using one-hot encoding
X_encoded = pd.get_dummies(X)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Train the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
rmse = np.sqrt(mse)
print("Root Mean Squared Error:", rmse)
```
